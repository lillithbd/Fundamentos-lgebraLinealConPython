```python
#Gráficos de manera interactiva debajo de las celdas
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
```

```python
#Sistema de ecuaciones sin soluciones
#No hay ningún punto de x que haga ciertas estas ecuaciones
#las 3 debieran cruzarse en un sólo punto
#sobredeterminado: tenemos más ecuaciones que variables

x = np.arange(-6, 6)

y_1 = 3*x+5
y_2 = -x+3
y_3 = 2*x + 1
```

```python
#Grafica
plt.figure()
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")
plt.show()
```

```python
#Sistema de ecuaciones con una solución
x = np.arange(-6,6)
#Determinado = igual número de ecuaciones que variables

y_2 = -1*x+3
y_3 = 2*x+1
```

```python
plt.figure()
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")
plt.show()
```

```python
#Infinitas soluciones
#Cualquier valor que tome X va a ser solución. 
#Indeterminado = Cuando me quedo con una ecuación pero tiene dos variables y tiene infinitas soluciones 
x = np.arange(-6,6)

y_3 = 2*x+1
```

```python
plt.figure()

plt.plot(x, y_3)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")
plt.show()
```
